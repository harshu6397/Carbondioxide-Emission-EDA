```python
# Importing necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataset in pandas dataframe
df = pd.read_csv('historical_emissions.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# Droping the unnecessary columns
df.drop(['Data source', 'Sector', 'Gas', 'Unit'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.sort_values(by='2018', ascending=False, inplace=True)
```

```python
df.head()
```

```python
data = df.T
```

```python
data.columns = df['Country'].values
```

```python
data.drop('Country', inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(10,10))
data['World'].plot()
plt.xlabel('Year')
plt.ylabel('MtCO₂e')
plt.title('Trend for Co2 emmissions from 1990 to 2018 in the World')
```

Obervation: The emmission of the carbondioxide is incresing very rapidedly.

```python
df1 = data.drop('World', axis=1)
```

```python
df1 = df1.iloc[:, :10]
```

```python
df1.head()
```

```python
df1.plot(figsize=(10,10))
plt.xlabel('Year')
plt.ylabel('MtCO₂e')
plt.title('Trend for Co2 emmissions from 1990 to 2018 in different countries')
```

Observation: China has the highest CO2 emmission rate, and the united sates and so on.

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Country', y='2018', data=df.iloc[1:][:10])
plt.xticks(rotation=90)
plt.title('Trend for Co2 emmissions in 2018')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Country', y='1990', data=df.iloc[1:][:10])
plt.xticks(rotation=90)
plt.title('Trend for Co2 emmissions in 1990')
plt.show()
```

```python
df['Change_in_emmission'] = df['2018'] - df['1990']
df.head()
```

```python
df.sort_values(by='Change_in_emmission', ascending=False)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Country', y='Change_in_emmission', data=df.sort_values(by='Change_in_emmission', ascending=False).iloc[1:][:10])
plt.xticks(rotation=90)
plt.title('Change in Trend for Co2 emmissions from 1990 to 2018')
plt.show()
```

Observation: There is a huge emmission rate in China and India observed easily from 1990 to 2018
